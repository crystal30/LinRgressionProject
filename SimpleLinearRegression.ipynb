# 一，简单的线性回归

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
y = np.array([2,3,5,8,9])
```

```python
#显然，这一组数据是有线性关系的,可以用现行回归来拟合。
x_mean = np.mean(x)
y_mean = np.mean(y)
molecular = 0.0
denominator = 0.0

for i in range(len(x)):
    molecular += (x[i] - x_mean)*(y[i] - y_mean)
    denominator += (x[i] - x_mean)**2
a = molecular/denominator
b = y_mean - a*x_mean
y_hat = a*x + b
```

```python
plt.scatter(x,y)
plt.plot(x,y_hat,color = 'r')
plt.show()
```

# 二， 用向量的形式来进行运算

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
y = np.array([2,3,5,8,9])
#显然，这一组数据是有线性关系的,可以用现行回归来拟合。

x_mean = np.mean(x)
y_mean = np.mean(y)

a = sum((x - x_mean)*(y - y_mean))/sum((x - x_mean)**2)
b = y_mean - a*x_mean
y_hat = a*x + b
```

```python
plt.scatter(x,y)
plt.plot(x,y_hat,color = 'r')
plt.show()
```

# 两个向量的对应元素相加再求和 等价于 两个向量做点乘
# 即sum((x - x_mean)*(y - y_mean)) == (x - x_mean).dot(y - y_mean)

```python
a = (x - x_mean).dot(y - y_mean)/(x - x_mean).dot(x - x_mean)
b = y_mean - a*x_mean
y_hat = a*x + b
plt.scatter(x,y)
plt.plot(x,y_hat,color = 'r')
plt.show()
```
